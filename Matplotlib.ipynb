```python
import matplotlib.pyplot as plt
%matplotlib inline


x = [2,4,7,9,12,15]
y = [2,5,8,9,12,13]
```

```python
#Plot the points as a line graph
plt.plot(x,y)


#Title the graph 'My Graph'
plt.title('My_Graph')

#Save the graph
plt.savefig('My_First_Graph.png')
```
